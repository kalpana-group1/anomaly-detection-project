```python
def parse_request(string):
    out = {}
    arr = string.split('/')
    out['page'] = arr[0]
    if len(arr) > 1:
        out['query'] = arr[1]
    else:
        out['query'] = ''
    return pd.Series(out)
```

```python
df['endpoint']=df['endpoint'].astype(str)
df = pd.concat([df,  df.endpoint.apply(parse_request)], axis=1)
```
